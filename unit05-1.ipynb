```python
# 2 14    ?

import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[] #     

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' : # 2/14  
            result.append(float(row[-1])) # result     
plt.plot(result,'hotpink')
plt.show()
```

```python
# 1983   , 2/14     

import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-') #   -   date  
        if 1983 <= int(date[0]) : #1983   
            if date[1] == '02' and date[2] ==  '14' : # 2 14 
                high.append(float(row[-1])) #   high  
                low.append(float(row[-2])) #   low  

plt.rc('font',family='Malgun Gothic') #
plt.title('    ') #
plt.rcParams['axes.unicode_minus'] = False #  (-)   .
plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
year=[]
average=[]

for row in data:
    if row[-3] != '':
        date = row[0].split('-')
        if date[1] == '02' and date[2] ==  '14' :
            year.append(float(date[0])) #  year  
            average.append(float(row[-3])) #   average  

#plt.rc('font',family='Malgun Gothic') #
plt.rcParams['axes.unicode_minus'] = False #  (-)  
plt.plot([year],[average],'hotpink')
plt.show()
```
